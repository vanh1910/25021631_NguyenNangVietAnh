```python
def maxOfTwo(a: int, b:int) -> int:
    return max(a,b)

a,b = map(int, input().split())
print(maxOfTwo(a,b))
```

```python
def swapUsingTuple(a,b) -> tuple[int,int]:
    return tuple((b,a))

a,b = map(int, input().split())
print(swapUsingTuple(a,b))
```

```python
import math
def isPrimeNumber(a: int) -> bool:
    if a == 1: 
        return False
    for i in range(2,math.isqrt(a)):
        if a % i == 0:
            return False
    return True

isPrimeNumber(a = int(input()))
```

```python
def isPerfectNumber(a: int) -> bool:
    sum = 0
    for i in range(1,a):
        if a % i == 0:
            sum += i
    if sum == a:
        return True
    return False

isPerfectNumber(a = int(input()))
```

```python
def firstKInList(arr: list, k: int) -> int:
    for i,x in enumerate(arr):
        if x == k:
            return i
        
    return -1
        
arr = []
arr = list(map(int,input().split()))
k = int(input())
print(firstKInList(arr,k))
```

```python
def factorial(x: int) -> int:
    ans = x
    for i in range(2,x):
        ans *= i
    return ans

factorial(x = int(input()))
```

```python
def calc(x: float, ch: str, y: float) -> float:
    if ch == "+":
        return x + y
    if ch == "-":
        return x - y
    if ch == "*":
        return x * y
    if ch == "/":
        return x / y
    
x,ch,y = input().split()
print(calc(float(x),ch,float(y)))
    
```

```python
def hammingDis(a: int, b: int) -> int:
    if (a>b): 
        a,b = b,a
    a = "{:1b}".format(a)
    b = "{:1b}".format(b)
    ans = 0
    for i in range (len(b) - len(a)):
        a = '0' + a
    
    for i in range (len(b)):
        if a[i] != b[i]:
            ans += 1
        

    return ans
a,b = map(int,input().split())
print(hammingDis(a,b))
```

```python
def totalDigit(a: int) -> int:
    a = str(a)
    ans = 0
    for i in a:
        ans += ord(i) - ord('0')
    
    return ans
print(totalDigit(a = int(input())))
```

```python
def isomorphism(a: str, b: str) -> bool:
    dict = {}
    for i in range (len(a)):
        if b[i] in dict:
            return False
        else: dict.update({b[i] : a[i]})

    return True

a = input()
b = input()
print(isomorphism(a,b))
```

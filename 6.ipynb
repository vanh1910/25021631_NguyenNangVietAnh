```python
dictionary = {}
a = list(map(int, input().split()))
for i in a: 
    if i not in dictionary:
        dictionary[i] = 1
    else:
        dictionary[i] += 1
a = [key for key in dictionary]
print(a)
```

```python
prefix_sum = []
a = list(map(int, input().split()))
for i in range(len(a)):
    if i == 0:
        prefix_sum.append(a[i])
    else:
        prefix_sum.append(prefix_sum[i-1] + a[i])
print(prefix_sum)
```

```python
tup = tuple(map(int, input().split()))
k = int(input())
arr = [x for x in tup]
for i in tup:
    arr.append(i)

for i in range(len(tup) - k, 2*len(tup) - k):
    print(arr[i], end = " ")
```

```python
inp = list(map(str,input().split()))
d = {}

for pair in inp:
    k,v = map(int,pair.split(":"))
    if k in d:
        d[k].append(v)
    else:
        d[k] = [v]

for k,v in d.items():
    print(k,v)
```

```python
d = {"possitives" : 0, "negatives": 0, "zeros": 0}
arr = list(map(int, input().split()))
for i in arr:
    if i > 0:
        d["possitives"] += 1
    elif i == 0:
        d["zeros"] += 1
    else:
        d["negatives"] += 1

print(d)
```
